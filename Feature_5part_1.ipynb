```python
import cv2
import numpy as np
from mlxtend.image import extract_face_landmarks
import math
import os.path
import csv
```

```python
def eye_aspect_ratio(eye):
	A = distance.euclidean(eye[1], eye[5])
	B = distance.euclidean(eye[2], eye[4])
	C = distance.euclidean(eye[0], eye[3])
	ear = (A + B) / (2.0 * C)
	return ear
```

```python
def mouth_aspect_ratio(mouth):
    A = distance.euclidean(mouth[14], mouth[18])
    C = distance.euclidean(mouth[12], mouth[16])
    mar = (A ) / (C)
    return mar
```

```python
def circularity(eye):
    A = distance.euclidean(eye[1], eye[4])
    radius  = A/2.0
    Area = math.pi * (radius ** 2)
    p = 0
    p += distance.euclidean(eye[0], eye[1])
    p += distance.euclidean(eye[1], eye[2])
    p += distance.euclidean(eye[2], eye[3])
    p += distance.euclidean(eye[3], eye[4])
    p += distance.euclidean(eye[4], eye[5])
    p += distance.euclidean(eye[5], eye[0])
    return 4 * math.pi * Area /(p**2)
```

```python
def mouth_over_eye(eye):
    ear = eye_aspect_ratio(eye)
    mar = mouth_aspect_ratio(eye)
    mouth_eye = mar/ear
    return mouth_eye
```

```python
def getFrame(sec):
    start = 60000
    vidcap.set(cv2.CAP_PROP_POS_MSEC, start + sec*500)
    hasFrames,image = vidcap.read()
    return hasFrames, image
```

```python
def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized
```

```python
# main program
from scipy.spatial import distance

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
data = []
labels = []
respondent = []
for j in range(61):
    for i in range(11):
        if j == 54:
            vidcap = cv2.VideoCapture('Fold5_part1/Fold5_part1/' + str(j) +'/' + str(i) + '.mp4')
        else:
            vidcap = cv2.VideoCapture('Fold5_part1/Fold5_part1/' + str(j) +'/' + str(i) + '.MOV')
        sec = 0
        frameRate = 1
        success, image  = getFrame(sec)
        count = 0
        
        while success and count < 400:
            #print("not detected")
            image = image_resize(image,width = 720 ,height = 1080)
            #
            if j < 54 and j > 48:
                if j == 49:
                    image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)
                elif j == 51:
                    sec
                else:
                    image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
                
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.3, 5)
            cv2.imshow('image',image)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
            if len(faces) == 0:

                sec = sec + frameRate
                sec = round(sec, 2)
                success, image = getFrame(sec)
                print("not detected")
            else:
                landmarks = extract_face_landmarks(gray)
                if landmarks is not None:
                    if sum(sum(landmarks)) != 0:
                        #print(landmarks)
                        count += 1
                        data.append(landmarks)
                        labels.append([i])
                        respondent.append([j])

                        sec = sec + frameRate
                        sec = round(sec, 2)
                        success, image = getFrame(sec)
                        print(count)

                else:  
                    sec = sec + frameRate
                    sec = round(sec, 2)
                    success, image = getFrame(sec)
                    print("not detected")
cv2.destroyAllWindows()
```
