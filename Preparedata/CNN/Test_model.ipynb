```python

from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
config = ConfigProto()
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score, f1_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import warnings
import tensorflow as tf
import os
import numpy as np
import pandas as pd
import cv2
import time
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from keras.models import load_model
from sklearn.metrics import classification_report
from random import shuffle
```

```python
model = load_model('CNN_Model_200.h5')
```

```python
def make_1080p():
    cap.set(3, 1920)
    cap.set(4, 1080)
```

```python
tf.debugging.set_log_device_placement(True)
counter = 0
x = 1
start_time = time.time()
cap = cv2.VideoCapture(0)
make_1080p()
text_fps=""
face_cascade = cv2.CascadeClassifier('C:/Users/nataw/Desktop/CEKMITL/Project4D/App/Project/Preparedata/haarcascade_frontalface_default.xml')
while True:
    counter+=1
    ret,frame = cap.read()
    thickness = 1
    color = (0, 0, 255)
    fontScale = 1
    org = (0, 185)
    org_2 = (0, 30)
    font = cv2.FONT_HERSHEY_SIMPLEX
    if ret:
        #img.reshape(1,224,224,3)/255.
        faces = face_cascade.detectMultiScale(frame, 1.3, 5)
        if len(faces) == 0:
            #img = cv2.resize(frame,(224,224))
            #print("not detected")
            continue
        elif len(faces)> 0:
            
            face_crop = []
            for f in faces:
                x, y, w, h = [ v for v in f ]
                #cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,0), 3)
                # Define the region of interest in the image  
                face_crop.append(frame[y:y+h, x:x+w])
            img = cv2.resize(face_crop[0],(224,224))
            pred = model.predict_classes(img.reshape(1,224,224,3)/255.)
            if pred == 0:
                text = ''
            elif pred == 1:
                text = 'Sleepy'

            
            
            frame = cv2.putText(frame, text, org, font, fontScale,  
                         color, thickness, cv2.LINE_AA, False)
        print(time.time() - start_time)
        if (time.time() - start_time) > 1 :
            text_fps  = str(counter / (time.time() - start_time))
            counter = 0
            start_time = time.time()
        frame = cv2.putText(frame, text_fps, org_2, font, fontScale,  
                             color, thickness, cv2.LINE_AA, False)
        cv2.imshow("window_name", frame)
        if cv2.waitKey(1) & 0xFF == 27:
            break
        
        
        
    else:
        print("camera not found")
cv2.destroyAllWindows()
cap.release()
```
