```python
import cv2
import numpy as np
from mlxtend.image import extract_face_landmarks
import math
import os.path
import csv
```

```python
def getFrame(sec):
    start = 60000
    vidcap.set(cv2.CAP_PROP_POS_MSEC, start + sec*250)
    hasFrames,image = vidcap.read()
    return hasFrames, image
```

```python
def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized
```

```python
from scipy.spatial import distance

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

for i in range(0,6):
    vidcap = cv2.VideoCapture('E:/Drowsiness/video/' + str(i) + '.mp4')
    sec = 0
    frameRate = 1
    success, image  = getFrame(sec)
    count = 0
    while success and count < 400:
        image = image_resize(image,width = 720 ,height = 1080)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        if len(faces) == 0:
            sec = sec + frameRate
            sec = round(sec, 2)
            success, image = getFrame(sec)
            print("not detected")
        else:
            
            face_crop = []
            count_face = 0
            for f in faces:
                x, y, w, h = [ v for v in f ]
                face_crop.append(image[y:y+h, x:x+w])
                img_name = str(count)+'_'+str(i)+'_'+str(count_face)
                cv2.imwrite('E:/Drowsiness/video/img/'+img_name+'.jpg', face_crop[-1])
                count_face+=1
            
            count += 1

            
            sec = sec + frameRate
            sec = round(sec, 2)
            success, image = getFrame(sec)
            print(count)

cv2.destroyAllWindows()
```
